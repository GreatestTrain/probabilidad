* **Distribución _gaussiana_ o distribución normal teórica**
    
    $$ P(X) = \frac{1}{\sigma\sqrt{2\pi}}exp\left[-\frac{1}{2}\left(\frac{X - \mu}{\sigma}\right)^2\right] $$
    + $exp$: $e^{x}$
    + $X$: Suceso
    + $\mu$: media de distribución
    + $\sigma$: desviación estándar de la distribución


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
## (Forma 1) Definición de la función de distribución gaussiana
def gaussian(x, mu, sigma):
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5*pow((x-mu)/sigma, 2))
```

El método <kbd>.pdf</kbd> permite calcular la probabilidad con cierto numero de intentos brindando la media y la desviación estándar de la distribución.

```python
## (Forma 2) Uso de scipy normç
dist = norm(0.0, 1)
dist.pdf(1)
```

```python
fig1, ax1 = plt.subplots()
x1 = np.arange(-4,4, 0.1)
# y1 = gaussian(x1, 0.0, 1)
y1 = [dist.pdf(i) for i in x1]
ax1.plot(x1, y1, color='purple', linestyle='--')
plt.show()
```

```python
## Distribucuón acumulada
fig2, ax2 = plt.subplots()
x2 = np.arange(-4,4, 0.1)
# y1 = gaussian(x1, 0.0, 1)
y2 = [dist.cdf(i) for i in x2]
ax2.plot(x2, y2, color='green', linestyle='--')
plt.show()
```

#### **Distribución _gaussiana_**

```python
df = pd.read_excel('./resources/seattle_central.xls')
arr = df['Normally Distributed Housefly Wing Lengths'].values[4:]
values , dist_values = np.unique(arr, return_counts=True)
```

```python
fig3, ax3 = plt.subplots()
ax3.bar(values, dist_values/len(arr), color='g')
plt.show()
```

```python
## Cálculo de mu y sigma
mu = arr.mean()
sigma = arr.std()
```

```python
# norm(mu, sigma)
dist = norm(mu, sigma)
```

```python
x4 = np.arange(30,60,0.1)
y4 = [dist.pdf(i) for i in x4]
fig4, ax4 = plt.subplots()
ax4.plot(x4, y4, color='r', linestyle='-.')
ax4.bar(values, dist_values/len(arr), color='g')
plt.show()
```
